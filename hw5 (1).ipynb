# data load

```python
import pandas as pd

# 데이터셋 로드
data = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temp': [30, 31, 28, 23, 10, 13, 14, 25, 15, 22, 19, 20, 33, 18],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'S-Wind': [1, 4, 2, 1, 2, 3, 5, 2, 1, 0, 3, 3, 1, 3],
    'W-Wind': [3, 2, 5, 3, 2, 2, 1, 3, 2, 4, 1, 2, 2, 3],
    'Play': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)
print(df)
```

# 1. Outlook=Overcast, Temp=25, Humidity=Normal, S-Wind=3, W-Wind=2

```python
import numpy as np
from math import comb
#1. Outlook=Overcast, Temp=25, Humidity=Normal, S-Wind=3, W-Wind=2

yes_df = df[df['Play'] == 'Yes']
no_df = df[df['Play'] == 'No']

#p(yes)
yes_count = yes_df.shape[0]
total_count = df.shape[0]
probability_yes = yes_count / total_count

# p(no)
no_count = no_df.shape[0]
total_count = df.shape[0]
probability_no = no_count / total_count

##############################yes#############################
#p (overcast|yes)
overcast_count = yes_df[yes_df['Outlook'] == 'Overcast'].shape[0]
yes_count = yes_df.shape[0]
probability_overcast = overcast_count / yes_count

#p(temp=25|yes) -use gausian distribution
mu = np.mean(yes_df['Temp']) #mean
sigma = np.std(yes_df['Temp']) #std_dev
x=25
probability_T = (1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(1/2)*((x - mu) / sigma) ** 2)) #gausian distribution

#p(normal|yes)
normal_count = yes_df[yes_df['Humidity'] == 'Normal'].shape[0]
yes_count = yes_df.shape[0]
probability_normal = normal_count / yes_count

#p(sw =3,ww=2|yes)
sw = 3
ww = 2

  #5C3
n = sw+ww
k = sw
com = comb(n, k)
  #p(sw|yes)
probability_sw = yes_df['S-Wind'].sum()/(yes_df['S-Wind'].sum() + yes_df['W-Wind'].sum())
  #p(ww|yes)
probability_ww = yes_df['W-Wind'].sum()/(yes_df['S-Wind'].sum() + yes_df['W-Wind'].sum())
probability_wind = com * (probability_sw ** sw) * (probability_ww ** ww)

pred_yes = probability_yes * probability_overcast * probability_T * probability_normal * probability_wind
print('pred_yes :',pred_yes)

##############################No#############################
#p (overcast|no) : Laplace smoothing
overcast_count = no_df[no_df['Outlook'] == 'Overcast'].shape[0] + 1
no_count = no_df.shape[0]
la_no_count= no_count + len(no_df['Outlook'].unique())
probability_overcast = overcast_count / la_no_count


#p(temp=25|no) -use gausian distribution
mu = np.mean(no_df['Temp']) #mean
sigma = np.std(no_df['Temp']) #std_dev
x=25
probability_T = (1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(1/2)*((x - mu) / sigma) ** 2)) #gausian distribution

#p(normal|no)
normal_count = no_df[no_df['Humidity'] == 'Normal'].shape[0]
probability_normal = normal_count / no_count

#p(sw =3,ww=2|no)
sw = 3
ww = 2

  #5C3
n = sw+ww
k = sw
com = comb(n, k)
  #p(sw|no)
probability_sw = no_df['S-Wind'].sum()/(no_df['S-Wind'].sum() + no_df['W-Wind'].sum())
  #p(ww|no)
probability_ww = no_df['W-Wind'].sum()/(no_df['S-Wind'].sum() + no_df['W-Wind'].sum())
probability_wind = com * (probability_sw ** sw) * (probability_ww ** ww)

pred_no = probability_no * probability_overcast * probability_T * probability_normal * probability_wind
print('pred_no :',pred_no)

```
