```python
import numpy as np

#초기 w0,w1 랜덤하게 선택
w0 = np.random.rand()
w1 = np.random.rand()

# 학습률: 하이퍼 파라미터
lr = 0.1

# 반복 횟수
t = 0

# stop point 설정
stop_point = 100

# stop point를 만족할 때까지 반복
while True:

    new_w0 = w0 - lr * (4 * w1 + 6 * w0 - 6)
    new_w1 = w1 - lr * (4 * w1 + 4 * w0 - 6)

    # w 업데이트
    w0 = new_w0
    w1 = new_w1
    #print("t="+str(t)+": w0 =", w0, "w1 =", w1)


    # 반복 횟수 업데이트
    t += 1

    # 정지 조건 확인
    if t >= stop_point:
        break

# 최종 결과 출력
print("최종 해(solution): w0 =", str(w0), "w1 =", str(w1))
```

```python
import math, sys, random
data = [(-1,1),(0,1),(1,1),(1,0)]

def E_w0(x,y,w0,w1,w2):
    return -2*(y-(w2*x+math.cos(w1)*x+w0))

def E_w1(x,y,w0,w1,w2):
    return -2*(-math.sin(w1)*x)*(y-(w2*x+math.cos(w1)*x+w0))

def E_w2(x,y,w0,w1,w2):
    return -2*x*(y-(w2*x+math.cos(w1)*x+w0))

# w0, w1, w2 = random.random(), random.random(), random.random()
w0, w1, w2 = 1,1,1
lr = 0.01
t = 0
threshold = sys.float_info.epsilon
max_t = 100000

while True:
    t += 1
    g0, g1, g2 = 0,0,0
    prev_w0, prev_w1, prev_w2 = w0, w1, w2
    for x,y in data:
        g0 += E_w0(x,y,w0,w1,w2)
        g1 += E_w1(x,y,w0,w1,w2)
        g2 += E_w2(x,y,w0,w1,w2)

    w0 = w0  - lr * g0
    w1 = w1 - lr * g1
    w2 = w2 - lr * g2

    if t > max_t or max(abs(prev_w0 - w0), abs(prev_w1 - w1), abs(prev_w2 - w2)) < threshold:
        break

print("w0:",w0," w1:",w1," w2:",w2)
```
