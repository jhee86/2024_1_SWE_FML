#1-2 Naïve Bayesian Classifier : Build a decision tree by ID3


```python
import math

def cal_Entropy(T, F):
    total = T + F
    if total == 0:
        return 0  # 총합이 0일 경우 엔트로피는 0

    # T와 F에 대한 확률 계산
    P_T = T / total
    P_F = F / total

    # P_T와 P_F가 0인 경우 처리
    if P_T == 0:
        entropy_T = 0
    else:
        entropy_T = - P_T * math.log2(P_T)

    if P_F == 0:
        entropy_F = 0
    else:
        entropy_F = - P_F * math.log2(P_F)

    # 엔트로피 계산 및 반환
    E = entropy_T + entropy_F
    return E
```

```python
def gain(E_parent,E_child):
  return E_parent - E_child
```

```python
# Data
import pandas as pd

# 데이터 정의
outlook = ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain']
temp = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']
humidity = ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High']
wind = ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Strong']
play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No']

# 데이터 프레임 생성
df = pd.DataFrame({
    'outlook': outlook,
    'temp': temp,
    'humidity': humidity,
    'wind': wind,
    'play': play
})

# 데이터 프레임 출력
print(df)

```
